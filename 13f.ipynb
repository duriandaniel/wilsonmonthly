```python
import pandas as pd
import requests

url = "https://www.sec.gov/Archives/edgar/data/1422849/000142284825000151/form13fInfoTable.xml"

headers = {
    "User-Agent": "Your Name Contact <daniel.li.sydney@gmail.com>",  # SEC requires real contact info
    "Accept-Encoding": "gzip, deflate",
    "Host": "www.sec.gov"
}

r = requests.get(url, headers=headers)
r.raise_for_status()  # throws if blocked

df = pd.read_xml(r.content)
print(df.head())
```
